# 3.3 Leapfrog Method and 1D Harmonic Oscillator

```python
import src.leapfrog as lf
import numpy as np
```

```python
k = 2
m = 1

time = 2 * np.pi
timesteps = 1000

dict_x, dict_v = {}, {}

for k in [1, 2, 5, 10]:
    Lx, Lv = lf.harmonic_oscillator(k, m, time, timesteps)

    dict_x[k] = Lx
    dict_v[k] = Lv

Lt = np.linspace(0, time, timesteps+1)

lf.visualise_harmonic_oscillator(Lt, dict_x, dict_v)
```
